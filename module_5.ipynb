```python
class phone:
    categore="Electronics"
#constructor
    def __init__(self,model,battery,camera):
        self.model=model
        self.battery=battery
        self.camera=camera

apple=phone("iphone 17", "5000 mAh", "108 MP")
print(type(apple))
print(apple.categore)
class smartphone(phone):
    def __init__(self, model, battery, camera, ram):
        super().__init__(model, battery, camera)  # parent constructor
        self.ram = ram
pro = smartphone("Samsung S23", "4500 mAh", "200 MP", "16 GB")

print(pro.model)
print(pro.battery)
print(pro.camera)
print(pro.ram)
```
