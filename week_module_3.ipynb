```python
from functools import reduce
#line count file
with open("D:/rabbi/PHITRON ML/test.txt", "r") as file:
    string=list(file.readlines())
    total_line=len(string)
    print(total_line)
# line by line word count
number_of_words=list(map(lambda x: len(x.split()), string))
print(number_of_words)
#total wrord count
total_word=sum(number_of_words)
print(total_word)
totalwords=reduce(lambda x,y: x+y, number_of_words)
print(totalwords)

#cleaning process
#1.new line delete
string1=list(map(str.strip, string))
print(string1)
#2.space delete
string2=list(map(lambda x: x.replace(" ", ""), string1))
print(string2)
#3.careter delete
string3=list(map(lambda x: len(x), string2))
print(sum(string3))
with open("D:/rabbi/PHITRON ML/conter_of_string.txt", "w") as file:
    file.write(f"Total line: {total_line}\n")
    file.write(f"Total word: {total_word}\n")
    file.write(f"Total carecter: {sum(string3)}\n")
```
