```python
def adder(a,b):
    return a+b
num=(input("Enter first number: "))
number=num.split()
x=int(number[0])
y=int(number[1])

result=adder(x,y)
print(result)
```

```python
def sum(a,b,c):
    print(f"{a},{c},{b}")
sum(10,30,20)
```

```python
def adder(*args):
    sum=0
    for i in args:
        sum=sum+i
    return sum
print(adder(10,20,30,40))
```

```python
def adder(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}={value}")
adder(name="rabbi",age=22,city="dhaka")
```
