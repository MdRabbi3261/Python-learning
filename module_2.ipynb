```python
inp=input()
number=inp.split()
x=int(number[0])
y=int(number[1])
z=int(number[2])

if x>=y and x>=z:
    mx=x
elif y>=z and y>=x:
    mx=y
else:
    mx=z

if x<=y and x<=z:
    mn=x
elif y<=z and y<=x:
    mn=y
else:    mn=z

print(str(mn) + " " + str(mx))
```

```python
inp=int(input())
inp1=input()
nu=inp1.split()

ev=0;od=0;pos=0;neg=0
for i in range(1,inp+1):
    num=int(nu[i-1])
    if(num%2==0):
        ev=ev+1
    else :
        od=od+1


    if (num>0):
        pos=pos+1
    elif (num<0):
        neg=neg+1

print("Even: "+str(ev))
print("Odd: "+str(od))
print("Positive: "+str(pos))
print("Negative: "+str(neg))


```
